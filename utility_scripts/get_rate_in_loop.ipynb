```python
import os
import numpy as np
import pandas as pd
from scipy.signal import hilbert
from scipy.signal import butter, lfilter, freqz
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
def get_signal(Path):
    envs = []
    files = os.listdir(Path)
    for f in files:
        if f.endswith('.wav'):
            #print(f)
            signal, sr = librosa.load(Path+'\\'+f)
            #t = np.arange(signal.size)/sr
            #print(t)
            #librosa.display.waveshow(signal, sr=sr)
            #print(sr, signal)
            analytic_signal = hilbert(signal)
            amplitude_envelope = np.abs(analytic_signal)
            envs.append(amplitude_envelope)
    return envs    
```

```python
def butter_lowpass_filter(data, cutoff, fs, order=5):
    all_peaks = []
    #print(cutoff)
    b, a = butter(5, 10, 'low', analog=False, fs=22050)
    y = lfilter(b, a, data)
    #print(y)
    peaks, _ = find_peaks(y, height=0.01)
    #print(peaks)
    #plt.plot(y)
    #plt.plot(peaks, y[peaks], "x")
    #plt.plot(np.zeros_like(y), "--", color="gray")
    #plt.show()
    all_peaks.append(len(peaks))
    return all_peaks
```

```python
C = get_signal('C:\\Users\\varya\\master_thesis\\out_C1_babble')
```

```python
len(C)
```

```python
times = []
fin_nested = []
for env in C:
    res = butter_lowpass_filter(env, 10, 220050, order=5)
    fin_nested.append(res)
    time = len(env)/22050
    times.append(time)
#print(times)
fin = [item for sublist in fin_nested for item in sublist]
print(fin)
```

```python
rates = [a/b for a, b in zip(fin, times)]
rates
#np.mean(rates)
```

```python
df = pd.DataFrame({"Noise": "Babble", "Rate": rates})
df.insert(0, "Talker", "C1")
#df
df.to_csv('C1_babble_rates.csv', encoding='utf-8', index=False)
```
